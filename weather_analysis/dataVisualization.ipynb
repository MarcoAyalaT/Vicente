```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df_atacama = pd.read_csv('atacama_hourly_jan2024.csv', skiprows=11)
df_quintay = pd.read_csv('quintay_hourly_jan2024.csv', skiprows=11)
df_unabcas = pd.read_csv('unabcas_hourly_jan2024.csv', skiprows=11)
```

```python
plt.rc('text', usetex=False)

plt.plot(df_atacama.iloc[0:23*3]['ALLSKY_SFC_SW_DWN'], ls='--', marker='x', color='r', label='Atacama')
plt.plot(df_quintay.iloc[0:23*3]['ALLSKY_SFC_SW_DWN'], ls='--', marker='x', color='g', label='Quintay')
plt.plot(df_unabcas.iloc[0:23*3]['ALLSKY_SFC_SW_DWN'], ls='--', marker='x', color='navy', label='Casona')
plt.xlabel('Hour of the day')
plt.ylabel('Solar Irradiance $[W/m^2]$')
plt.grid(ls='--')
plt.title('Solar Irradiance, Jan 1st-3rd 2024')
plt.legend(title='Place:')
plt.show()
```

```python
plt.plot(df_atacama.iloc[0:23*3]['T2M'], ls='--', marker='x', color='r', label='Atacama')
plt.plot(df_quintay.iloc[0:23*3]['T2M'], ls='--', marker='x', color='g', label='Quintay')
plt.plot(df_unabcas.iloc[0:23*3]['T2M'], ls='--', marker='x', color='navy', label='Casona')
plt.xlabel('Hour of the day')
plt.ylabel('Temperature @ 2 meters $[C]$')
plt.grid(ls='--')
plt.title('Temperature, Jan 1st-3rd 2024')
plt.legend(title='Place:')
plt.show()
```
